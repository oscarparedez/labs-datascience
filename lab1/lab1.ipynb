```python
# Se eliminan variables que aportan muy poco valor al analisis del conjunto de datos
data.drop(['Id', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'BsmtHalfBath', 'LowQualFinSF', 'BsmtFinSF2'],axis=1,inplace=True)
```

```python
data.dtypes
columns = data.columns
cualitativas, cuantitativas, cualitativasEspeciales = [], [], []
for i, t in enumerate(data.dtypes):
    t_str = str(t)
    column = columns[i]
    if t_str == 'int64' or t_str == 'float64' :
        if len(pd.unique(data[column])) < 20:
            cualitativasEspeciales.append(column)
        cuantitativas.append(column)
    else:
        cualitativas.append(column)
```

```python
import pandas as pd
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
data_test = data[cuantitativas]
data_test.dropna(inplace=True)
chi_square_value,p_value=calculate_bartlett_sphericity(data_test)
chi_square_value, p_value
```

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(data_test)
kmo_all,kmo_model
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
escalador = StandardScaler()
escalador.fit(data_test)
datos_escalados = escalador.transform(data_test)
pca = PCA(n_components=4)
pca.fit(datos_escalados)
x_pca = pca.transform(datos_escalados)
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=data_test['SalePrice'],cmap='plasma')
plt.legend(data_test[cuantitativas], loc = "upper right")
plt.xlabel('Primer componente principal')
plt.ylabel('Segundo componente principal')
```

```python
pca.components_
df_comp = pd.DataFrame(pca.components_,columns=cuantitativas)
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap="viridis",)  # probar con cmap="plasma", para otros colores
```

```python
categoricas = ['Alley', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'OverallQual', 'OverallCond', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']
data_categoricas = data[categoricas].sample(frac=0.1)
data_categoricas = data_categoricas.astype(object)
```

```python
# Se van a convertir los datos a una lista de listas
print(data_categoricas.shape)
records = []
for i in range(0, 146):
    records.append([str(data_categoricas.values[i,j]) for j in range(0, 6)])
```

```python
# El mÃ­nimo de cobertura o soporte es 20% y el mÃ­nimo de confianza es 70%
reglas_asociacion = apriori(records,min_support=0.1, min_confidence = 0.7)
reglas = list(reglas_asociacion)
```
